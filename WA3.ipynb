## 1. `while` Loop That Terminates When User Types "exit"

```python
while True:
    user_input = input("Enter something (or type 'exit' to quit): ")
    if user_input.strip().lower() == "exit":
        print("Exiting loop...")
        break
    print(f"You entered: {user_input}")

```

## 2. Print Students Who Passed (>35) from a Dictionary

```python
students_marks = {
    "Alice": 85,
    "Bob": 28,
    "Charlie": 45,
    "David": 32,
    "Eva": 92
}

print("Passed Students:")
for name, marks in students_marks.items():
    if marks > 35:
        print(f"- {name}: {marks}")

```

## 3. List Comprehension to Square All Numbers in a List

```python
numbers = [1, 2, 3, 4, 5, 6]
squared_numbers = [x**2 for x in numbers]

print(f"Original: {numbers}")
print(f"Squared:  {squared_numbers}")  # Output: [1, 4, 9, 16, 25, 36]

```

## 4. Merge Two Dictionaries in Python 3.9+

Python 3.9 introduced the union operator (`|`) for merging dictionaries cleanly.

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 99, "c": 4}  # 'b' will be overwritten by dict2's value

# Using the merge operator '|'
merged_dict = dict1 | dict2

print(merged_dict)  # Output: {'a': 1, 'b': 99, 'c': 4}

```

## 5. Frequency of Each Character in a String

```python
text = "hello world"
char_frequency = {}

for char in text:
    char_frequency[char] = char_frequency.get(char, 0) + 1

for char, count in char_frequency.items():
    print(f"'{char}': {count}")

```

## 6. Demonstrate `enumerate()` in a `for` Loop

The `enumerate()` function yields both the index and the value during iteration.

```python
fruits = ["apple", "banana", "cherry", "date"]

for index, fruit in enumerate(fruits, start=1):
    print(f"Index {index}: {fruit}")

```

## 7. Intersection of Two Sets

You can use either the `.intersection()` method or the `&` operator.

```python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# Using the & operator
intersection_set = set_a & set_b

print("Intersection:", intersection_set)  # Output: {4, 5}

```

## 8. Combine Two Lists into a Dictionary Using `zip()`

```python
keys = ["name", "age", "city"]
values = ["Alice", 25, "New York"]

# Combine lists using zip() and cast to dict
combined_dict = dict(zip(keys, values))

print(combined_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}

```

## 9. $3 \times 3$ Matrix Using Nested Lists & Access Middle Element

```python
# 3x3 Matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# The middle element is at row index 1, column index 1
middle_element = matrix[1][1]

print(f"Middle element: {middle_element}")  # Output: 5

```

## 10. Sort a Dictionary by Its Values

Use the built-in `sorted()` function passing a custom key (`lambda item: item[1]`).

```python
scores = {"Alice": 88, "Bob": 75, "Charlie": 95, "David": 60}

# Sort dictionary by value (ascending order)
sorted_scores = dict(sorted(scores.items(), key=lambda item: item[1]))

print(sorted_scores)  
# Output: {'David': 60, 'Bob': 75, 'Alice': 88, 'Charlie': 95}

```
