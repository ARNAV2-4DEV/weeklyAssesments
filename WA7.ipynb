## 1. Class `Car` with Attributes and Display Method

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_details(self):
        print(f"Car Details: {self.year} {self.make} {self.model}")

# Usage
my_car = Car("Toyota", "Camry", 2022)
my_car.display_details()

```

## 2. Purpose of the `__init__` Method

The `__init__` method is Python's **constructor** method. It automatically runs whenever a new object/instance of a class is created. Its main purpose is to **initialize instance attributes** with initial values.

```python
class Person:
    def __init__(self, name):
        self.name = name  # Initializing attribute upon instantiation

```

## 3. Subclass `ElectricCar` Inheriting from `Car`

```python
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

class ElectricCar(Car):
    def __init__(self, make, model, battery_capacity):
        super().__init__(make, model)  # Call parent class constructor
        self.battery_capacity = battery_capacity

# Usage
tesla = ElectricCar("Tesla", "Model 3", "75 kWh")
print(f"{tesla.make} {tesla.model} - Battery: {tesla.battery_capacity}")

```

## 4. Demonstrate "Method Overriding" in Inheritance

Method overriding happens when a child/subclass provides its own implementation of a method that is already defined in its parent class.

```python
class Animal:
    def make_sound(self):
        print("Some generic sound")

class Dog(Animal):
    def make_sound(self):  # Overriding parent method
        print("Woof! Woof!")

dog = Dog()
dog.make_sound()  # Output: Woof! Woof!

```

## 5. Encapsulation with Private Attributes (`__`)

Encapsulation restricts direct access to specific object attributes. Prefixing an attribute with double underscores (`__`) triggers name mangling, making it private to the class.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute

    def get_balance(self):
        return self.__balance

account = BankAccount("Alice", 1000)
print(account.get_balance())  # Output: 1000
# print(account.__balance)     # Raises AttributeError

```

## 6. Use `@property` for Getter and Setter Methods

```python
class Student:
    def __init__(self, score):
        self._score = score  # Protected attribute convention

    @property
    def score(self):  # Getter
        return self._score

    @score.setter
    def score(self, new_score):  # Setter
        if 0 <= new_score <= 100:
            self._score = new_score
        else:
            print("Invalid score! Must be between 0 and 100.")

s = Student(85)
print(s.score)     # Output: 85
s.score = 95       # Invokes setter
print(s.score)     # Output: 95

```

## 7. Multiple Inheritance with Code Structure

Multiple inheritance occurs when a single class inherits attributes and methods from more than one parent class.

```python
class Engine:
    def start_engine(self):
        print("Engine started")

class Radio:
    def play_music(self):
        print("Playing music")

# Car inherits from both Engine and Radio
class SmartCar(Engine, Radio):
    pass

car = SmartCar()
car.start_engine()  # From Engine
car.play_music()    # From Radio

```

## 8. Class Method Using `@classmethod` Decorator

Class methods bind to the class itself (`cls`) rather than instances (`self`). They are often used as alternative constructors.

```python
class Person:
    species = "Homo sapiens"

    def __init__(self, name):
        self.name = name

    @classmethod
    def get_species(cls):
        return cls.species

print(Person.get_species())  # Output: Homo sapiens

```

---

## 9. Class Method vs. Static Method

| Feature | `@classmethod` | `@staticmethod` |
| --- | --- | --- |
| **First Parameter** | Receives `cls` (the class) implicitly. | Receives no implicit first parameter (`self` or `cls`). |
| **State Access** | Can modify class state that applies across all instances. | Cannot access or modify class or instance state directly. |
| **Primary Purpose** | Factory methods or modifying class-level data. | Utility/helper methods that are self-contained. |

## 10. `try-except-finally` Block for `ZeroDivisionError`

```python
try:
    num1 = 10
    num2 = 0
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
finally:
    print("Execution completed (finally block always runs).")

```
