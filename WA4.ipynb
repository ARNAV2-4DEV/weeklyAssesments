## 1. Explain `break`, `continue`, and `pass` Statements

* **`break`:** Immediately exits the current loop entirely.
* **`continue`:** Skips the rest of the current iteration and jumps to the next cycle of the loop.
* **`pass`:** A null statement that acts as a placeholder when code is syntactically required but no action is needed.

```python
# break: stops when num == 3
for num in range(1, 6):
    if num == 3:
        break
    print(f"break loop: {num}")  # Output: 1, 2

# continue: skips num == 3
for num in range(1, 6):
    if num == 3:
        continue
    print(f"continue loop: {num}")  # Output: 1, 2, 4, 5

# pass: placeholder
for num in range(1, 4):
    if num == 2:
        pass  # TODO: add logic later

```

## 2. Convert Celsius to Fahrenheit Using `map()`

Formula: $F = (C \times 9/5) + 32$

```python
celsius_temps = [0, 20, 37, 100]

# Using map with a lambda function
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print("Celsius:   ", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)  # Output: [32.0, 68.0, 98.6, 212.0]

```

## 3. Find the Second Smallest Element in a List

```python
numbers = [12, 45, 2, 19, 2, 8, 45]

# Remove duplicates using set(), sort the values, then access index 1
unique_sorted = sorted(set(numbers))

if len(unique_sorted) > 1:
    second_smallest = unique_sorted[1]
    print(f"Second smallest: {second_smallest}")  # Output: 8
else:
    print("List does not have enough unique elements.")

```

## 4. Check if a Year is a Leap Year

A year is a leap year if it is divisible by 4, except for century years (ending in 00), which must also be divisible by 400.

```python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Test
year = 2024
print(f"Is {year} a leap year?", is_leap_year(year))  # Output: True

```

## 5. Reverse a Dictionary (Keys Become Values & Vice Versa)

```python
original_dict = {"a": 1, "b": 2, "c": 3}

# Using dictionary comprehension
reversed_dict = {value: key for key, value in original_dict.items()}

print(reversed_dict)  # Output: {1: 'a', 2: 'b', 3: 'c'}

```

## 6. Demonstrate `any()` and `all()` on a List of Booleans

* **`any()`:** Returns `True` if **at least one** element evaluates to `True`.
* **`all()`:** Returns `True` if **every** element evaluates to `True`.

```python
bool_list_1 = [True, False, False]
bool_list_2 = [True, True, True]

print(any(bool_list_1))  # Output: True  (at least one True)
print(all(bool_list_1))  # Output: False (not all are True)
print(all(bool_list_2))  # Output: True  (all are True)

```

## 7. Flatten a Nested List

```python
nested_list = [[1, 2], [3, 4], [5, 6]]

# Flattening using list comprehension
flattened = [item for sublist in nested_list for item in sublist]

print(flattened)  # Output: [1, 2, 3, 4, 5, 6]

```

## 8. Difference Between `append()` and `extend()`

* **`append(object)`:** Adds its single argument as a **single element** to the end of the list.
* **`extend(iterable)`:** Iterates through its argument, adding **each element** individually to the end of the list.

```python
list1 = [1, 2]
list1.append([3, 4])
print("After append:", list1)  # Output: [1, 2, [3, 4]]

list2 = [1, 2]
list2.extend([3, 4])
print("After extend:", list2)  # Output: [1, 2, 3, 4]

```

## 9. Find the Longest Word in a Sentence

```python
sentence = "Python programming language is exceptionally versatile"

words = sentence.split()
longest_word = max(words, key=len)

print(f"Longest word: '{longest_word}' (Length: {len(longest_word)})")
# Output: 'exceptionally'

```

## 10. Find 3 Most Common Items Using `collections.Counter`

```python
from collections import Counter

items = ["apple", "banana", "apple", "cherry", "banana", "apple", "date", "banana", "elderberry"]

# Create Counter object and get top 3
counter = Counter(items)
most_common_three = counter.most_common(3)

print("3 Most Common Items:")
for item, count in most_common_three:
    print(f"- {item}: {count} times")

```
