## 1. Join a List of Strings into a Single Sentence

Use Python's built-in `str.join()` method, specifying a space `" "` (or any delimiter) as the separator.

```python
words = ["Python", "is", "an", "awesome", "programming", "language."]
sentence = " ".join(words)

print(sentence)  
# Output: "Python is an awesome programming language."

```

## 2. Remove Duplicates from a List Using a Set

Convert the list into a `set` (which automatically removes duplicate values) and then convert it back into a `list`.

```python
numbers = [1, 2, 2, 3, 4, 4, 4, 5]
unique_numbers = list(set(numbers))

print(unique_numbers)  # Output: [1, 2, 3, 4, 5] (order may vary in sets)

```

## 3. Check if a Specific Key Exists in a Dictionary

The most Pythonic way is to use the `in` operator.

```python
student = {"name": "Alex", "age": 21, "grade": "A"}

# Checking key existence
if "age" in student:
    print("Key 'age' exists!")
else:
    print("Key not found.")

```

## 4. Difference Between `is` and `==`

* `==` **(Equality Operator):** Checks if the **values** of two objects are equal.
* `is` **(Identity Operator):** Checks if two variables point to the exact same **memory location** (i.e., whether `id(a) == id(b)`).

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)  # True  (Their content/values are identical)
print(a is b)  # False (They are two distinct objects in memory)
print(a is c)  # True  (c points to the exact same object as a)

```

## 5. Simple Calculator Using `if-elif-else` Logic

```python
num1 = float(input("Enter first number: "))
op = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if op == '+':
    result = num1 + num2
elif op == '-':
    result = num1 - num2
elif op == '*':
    result = num1 * num2
elif op == '/':
    result = num1 / num2 if num2 != 0 else "Error: Division by zero"
else:
    result = "Invalid Operator"

print(f"Result: {result}")

```

## 6. Find the Largest Number in a List Without Using `max()`

```python
numbers = [14, 67, 3, 89, 42, 89, 12]

# Initialize with the first element
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print(f"The largest number is: {largest}")  # Output: 89

```

## 7. Demonstrate `.strip()` on a Messy Input String

The `.strip()` method removes leading and trailing whitespaces (spaces, tabs `\t`, newlines `\n`).

```python
messy_input = "   \n\t  Hello, World!   \t \n"
clean_input = messy_input.strip()

print(f"Original: '{messy_input}'")
print(f"Cleaned:  '{clean_input}'")  # Output: 'Hello, World!'

```

## 8. Convert Float to Integer & What Happens to Decimals

Convert a float using the `int()` function. Python **truncates** (drops) the decimal portion entirely—it does **not** round to the nearest integer.

```python
val = 9.876
converted = int(val)

print(f"Original: {val}")       # Output: 9.876
print(f"Converted: {converted}") # Output: 9

```

## 9. Multiplication Table of a Given Number

```python
num = int(input("Enter a number: "))

print(f"--- Multiplication Table for {num} ---")
for i in range(1, 11):
    print(f"{num} x {i:2d} = {num * i}")

```

## 10. Use of the `in` Keyword in Python

The `in` keyword serves two primary purposes:

1. **Membership Test:** Checks if a value exists within a sequence (string, list, tuple, set, dict).
2. **Iteration:** Used in `for` loops to iterate over elements in an iterable.

```python
# 1. Membership Test
text = "Python Programming"
print("Python" in text)  # Output: True

# 2. Iteration in Loop
colors = ["red", "green", "blue"]
for color in colors:
    print(color)

```
