## 1. Recursive Function for the Fibonacci Sequence

Returns the $n$-th Fibonacci number using recursion.

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

# Print first 6 Fibonacci numbers (0-indexed base)
for i in range(6):
    print(f"Fibonacci({i}) = {fibonacci(i)}")

```

## 2. Memoization and How It Helps Recursive Functions

**Memoization** is an optimization technique that caches (stores) the results of expensive function calls so that when the same inputs occur again, the function returns the cached result instead of recomputing it.

It optimizes recursion by turning exponential time complexity (like $O(2^n)$ in basic recursive Fibonacci) into linear time complexity ($O(n)$) by avoiding repeated redundant calculations.

```python
from functools import lru_cache

# Using Python's built-in memoization decorator
@lru_cache(maxsize=None)
def fast_fibonacci(n):
    if n <= 1:
        return n
    return fast_fibonacci(n - 1) + fast_fibonacci(n - 2)

print(fast_fibonacci(50))  # Extremely fast compared to standard recursion

```

## 3. Product of a List Using `functools.reduce()`

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# reduce applies the lambda sequentially across the list elements
product = reduce(lambda x, y: x * y, numbers)

print(f"Product: {product}")  # Output: 120

```

## 4. Function to Check if a Number is an "Armstrong Number"

An Armstrong number (narcissistic number) is a number that is equal to the sum of its own digits each raised to the power of the total number of digits (e.g., $153 = 1^3 + 5^3 + 3^3$).

```python
def is_armstrong(num):
    num_str = str(num)
    num_digits = len(num_str)
    
    total_sum = sum(int(digit) ** num_digits for digit in num_str)
    return total_sum == num

# Test
print(is_armstrong(153))  # True
print(is_armstrong(123))  # False

```

## 5. Closure That Remembers a Base Value and Adds It to an Input

```python
def make_adder(base_value):
    def add(x):
        return base_value + x  # Remembers base_value from outer scope
    return add

# Create an adder function with a fixed base of 10
add_ten = make_adder(10)

print(add_ten(5))   # Output: 15
print(add_ten(20))  # Output: 30

```

## 6. LEGB Rule in Python Scope

The **LEGB rule** defines the exact hierarchy Python follows when searching for variable names:

1. **L — Local:** Variables defined inside the current function.
2. **E — Enclosing:** Variables in any outer/enclosing functions (from inside to out).
3. **G — Global:** Variables defined at the top level of the module/file.
4. **B — Built-in:** Built-in names pre-assigned in Python (e.g., `print`, `len`, `sum`).

## 7. Convert Decimal to Binary Without `bin()`

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    
    binary_digits = []
    temp = n
    while temp > 0:
        binary_digits.append(str(temp % 2))
        temp = temp // 2
        
    return "".join(reversed(binary_digits))

# Test
print(decimal_to_binary(13))  # Output: "1101"

```

## 8. Generator Function producing Infinite Even Numbers

```python
def infinite_evens():
    num = 0
    while True:
        yield num
        num += 2

# Usage example (getting first 5 even numbers)
gen = infinite_evens()
for _ in range(5):
    print(next(gen), end=" ")  # Output: 0 2 4 6 8
print()

```

## 9. Passing a Function as an Argument to Another Function

In Python, functions are **first-class objects**, meaning they can be passed around as arguments just like integers or strings.

```python
def apply_operation(func, x, y):
    return func(x, y)

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

print(apply_operation(add, 4, 5))       # Output: 9
print(apply_operation(multiply, 4, 5))  # Output: 20

```

## 10. Validate Password (min 8 chars, 1 digit, 1 symbol)

```python
import string

def validate_password(password):
    if len(password) < 8:
        return False
    
    has_digit = any(char.isdigit() for char in password)
    has_symbol = any(char in string.punctuation for char in password)
    
    return has_digit and has_symbol

# Tests
print(validate_password("Pass123!"))  # True
print(validate_password("pass"))      # False (too short)
print(validate_password("Password"))  # False (missing digit and symbol)

```
